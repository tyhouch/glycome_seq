```python
import pandas as pd
import glob
```

```python
count = 0
glycosylated = []
files = glob.glob("/Users/tylerhouchin/Desktop/glycome_seq/lectin_database/*")
for file in files:
    lectin = open(file)
    lines = lectin.readlines()
    for l in lines:
        if l.find('CARBOHYD') != -1:
            glycosylated.append(file)
            break
print(len(glycosylated))
print(glycosylated[1])
```

```python
accessions = []
for file in glycosylated:
    lectin = open(file)
    lines = lectin.readlines()
    for l in lines:
        if 'AC   ' in l:
            acc = l
            accessions.append(acc)
print(accessions)
            
                
```
